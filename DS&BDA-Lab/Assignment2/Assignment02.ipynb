## 3)Using Z-Score

```python
z
```

```python
threshold = 0.18 
```

## 4)Using Inter Quantile Range(IQR)

```python
sorted_rscore = np.sort(df['Reading_Score'])
```

```python
sorted_rscore
```

```python
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
```

```python
"Q1:", q1, "Q3:", q3
```

```python
IQR = q3 - q1
```

```python
lwr_bound = q1 - (1.5 * IQR)
upr_bound = q3 + (1.5 * IQR)
"Lower Bound:", lwr_bound, "Upper Bound:", upr_bound
```

```python
r_outliers = []
for i in sorted_rscore:
    if i < lwr_bound or i > upr_bound:
        r_outliers.append(i)

"Outliers:", r_outliers
```
